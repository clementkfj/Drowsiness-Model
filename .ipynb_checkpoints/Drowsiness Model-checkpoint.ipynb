```python
import torch 
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
# device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
img = 'https://media.wired.com/photos/593256b42a990b06268a9e21/master/pass/traffic-jam-getty.jpg'
results = model(img) # passing the image into the YOLOV5
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))  # parsing the results into a numpy array and showing the photo
plt.show()
```

```python
plt.imshow(np.squeeze(results.render()))
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Use YOLO on frames
    results = model(frame)
    
    cv2.imshow('Webcam check', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid # images unique identifier
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images')
labels = ['awake', 'drowsy']
number_images = 20
```

```python
cap = cv2.VideoCapture(0)
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for image_num in range(number_images):
        
        print('Collecting images for {}'.format(label, number_images))
        
        ret, frame = cap.read()
        
        image_name = os.path.join(IMAGES_PATH, label + '.' + str(uuid.uuid1()) + '.jpg')
        cv2.imwrite(image_name, frame)
        cv2.imshow('Image collection', frame)
        time.sleep(2)
    if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
